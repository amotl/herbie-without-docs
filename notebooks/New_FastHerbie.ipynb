# New `FastHerbie()` Class
- uses multithreading to create many Herbie objects: `fh = FastHerbie(DATES)`.
- method for downloading those Herbie objects `fh.download()`
- method for opening that data with xarray `fh.xarray()`

```python
from herbie import FastHerbie
import pandas as pd
```

### HRRR

```python
%%time
# Create a range of dates
DATES = pd.date_range('2022-01-01', '2022-01-01 06:00', freq='1H')

# Create a range of forecast lead times
fxx = range(0,4)

# Make FastHerbie Object.
FH = FastHerbie(DATES, model='hrrr', fxx=fxx)
FH
```

```python
len(FH.objects)
```

```python
FH.download("TMP:2 m")
```

```python
# Download/Read Data with Xarray
ds = FH.xarray("TMP:2 m")
ds
```

### GFS

```python
DATES = pd.date_range('2022-01-01', '2022-01-02', freq='6H')

# Make Fast Herbie Object
fh = FastHerbie(DATES, model='gfs', fxx=[0])
fh.objects
```

```python
ds = fh.xarray(":TMP:2 m above ground")
ds
```

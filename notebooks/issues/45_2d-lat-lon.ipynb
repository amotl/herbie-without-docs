# HRRR model coordinates: [Discussion #45](https://github.com/blaylockbk/Herbie/discussions/45)

```python
from herbie.archive import Herbie
from toolbox.cartopy_tools import common_features, pc
import matplotlib.pyplot as plt
```

```python
ds = Herbie('2022-02-14', model='hrrr').xarray('TMP:2 m')
```

```python
ax = common_features('50m', dark=False, figsize=[10,10]).STATES().OCEAN().LAND().ax

ax.pcolormesh(ds.longitude, ds.latitude, ds.t2m, cmap='Spectral_r', transform=pc)

xs = ds.isel(y=slice(600,650))  # Get the 500th row cross-section

ax.pcolormesh(xs.longitude, xs.latitude, xs.t2m, cmap='Greys', transform=pc)

ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

ax.adjust_extent()
```

```python
ax = common_features('50m', crs=ds.herbie.crs, dark=False, figsize=[10,10]).STATES().OCEAN().LAND().ax

ax.pcolormesh(ds.longitude, ds.latitude, ds.t2m, cmap='Spectral_r', transform=pc)

xs = ds.isel(y=slice(600,650))  # Get the 500th row cross-section

ax.pcolormesh(xs.longitude, xs.latitude, xs.t2m, cmap='Greys', transform=pc)

ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)

ax.adjust_extent()
```
